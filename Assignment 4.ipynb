```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
religion = pd.read_csv('WRP_national.csv', usecols=[0,1,2,73])
religion = religion[religion['name'] == 'USA']
religion = religion.drop(religion.index[0])
religion = religion.drop(religion.index[12]).reset_index()
religion
```

```python
unemployment = pd.read_excel('cpsaat01.xlsx', skiprows=7, usecols=[0,9], names=['year', 'Unemployment Percent'], skipfooter=1)
years = list(np.arange(1950, 2010, 5))
unemployment = unemployment[unemployment['year'].isin(years)].reset_index()
unemployment['nonreligpct'] = religion['nonreligpct']
unemployment
```

```python
plt.figure()
plt.plot(unemployment['year'], unemployment['Unemployment Percent'],'-o',unemployment['year'], unemployment['nonreligpct'],'-o')
plt.legend(('Unemployment', 'Nonreligious'))
plt.ylabel('Percentage %')
```
